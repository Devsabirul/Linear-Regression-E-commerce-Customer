```python
df = df.drop(['Email', 'Address', 'Avatar'],axis=1)
```

```python
X = df.drop('Yearly Amount Spent', axis=1)
```

```python
X
```

```python
y = df['Yearly Amount Spent']
```

# Training Test Split 

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test, y_train,y_test = train_test_split(X,y, train_size=.70, random_state=51)
```

```python
X_test
```

# predict with Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
reg.fit(X_train,y_train)
```

```python
predict = reg.predict(X_test)
```

```python
y_test
```

```python
reg.score(X_test,y_test)
```

```python
df.head(2)
```

```python
df["Predict Yearly Amount Spent"] = reg.predict(X)
```

```python
df.head(5)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
print("Accuracy of training dataset:", reg.score(X_train,y_train))
print("Accuracy of test dataset:", reg.score(X_test,y_test))
```

# mean_squared_error , mean_absolute_error

```python
from sklearn.metrics import mean_squared_error , mean_absolute_error
```

```python
print('MAE = ', mean_absolute_error(y_test,predict))
```

```python
print('MSE =', mean_squared_error(y_test,predict))
```
